```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.head(5)
```

```python
features = ["study_hours_per_day", "attendance_percentage", "sleep_hours", "mental_health_rating", "part_time_job"]
```

```python
target = "exam_score"
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])
```

```python
df_model
```

```python
X = df_model[features]
```

```python
Y = df_model[target]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
len(Y_test)
```

```python
len(Y_train)
```

```python
models = {
    "Linear Regression": {
        "model": LinearRegression(),
        "params": {}   # no hyperparameters to tune
    },

    "Decision Tree Regressor": {
        "model": DecisionTreeRegressor(),
        "params": {"max_depth": [3, 5, 10], "min_samples_split": [2, 5]}
    },
    "Random Forest Regressor": {
        "model": RandomForestRegressor(),
        "params": {"n_estimators": [50, 100, 200], "max_depth": [10, 10, None]}
    }
}
```

```python
best_models = []
```

```python

for name, config in models.items():
  print(f"Training {name}")

  grio = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")
  grio.fit(X_train, Y_train)

  Y_pred = grio.predict(X_test)

  rmse =np.sqrt(mean_squared_error(Y_test, Y_pred))
  r2 = r2_score(Y_test, Y_pred)

  best_models.append({
      "model_name": name,
      "best_params": grio.best_params_,
      "rmse": rmse,
      "r2": r2
  })
```

```python
display(best_models)
```

```python
result_df = pd.DataFrame(best_models)
```

```python
result_df.sort_values(by="rmse")
```

```python
import joblib

best_row = result_df.sort_values(by="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row["model_name"]
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model = best_model_config["model"].set_params(**best_row["best_params"])
```

```python
final_model.fit(X,Y)
```

```python
final_model.predict
```

```python
joblib.dump(final_model, "best_model.pkl")
```
